# <center>Joining Dataframe</center>

Combining Multiple dataframe includes joining data with related field to add new columns,appending data with same field to add rows.<br>
* Append the rows for similar dataframe.
* Join the dataframe with single and multiple related field.<br>
There are two ways to combine dataframe<br>
--Appending and Joining--<br>
1. Appending - Stack the rows from multiple dataframe with the same column structure.
2. Joining - Add related columns from one dataframe to another,based on common values.


```python
import pandas as pd
```

```python
retail = pd.read_csv("retail_2016_2017.csv")
retail.head()
```

```python
automotive_df  =retail.query("family == 'AUTOMOTIVE'").head()
automotive_df.head()
```

```python
beauty_df = retail.query("family == 'BEAUTY'").head()
beauty_df =beauty_df.assign(new_col =100)
```

```python
beauty_df.head()
```

```python
pd.concat([automotive_df,beauty_df])
```

```python
pd.concat([automotive_df,beauty_df],ignore_index=True)
```

## Joining Dataframe

You can join dataframe with the merge() method<br>
* The dataframe must share atleast one column to match the values between them<br>
There are primary four types of join :<br>
1. Inner join
2. Left join 
3. Right join<br>
4. Outer join<br>
* Inner and left joins are most common used ,while right is rarely used.



```python
oil = pd.read_csv("oil.csv",parse_dates=["date"])
```

```python
transactions =pd.read_csv("transactions.csv",parse_dates=["date"])
```

```python
oil.info()
```

```python
transactions.info()
```

```python
transactions.merge(oil,how ="inner",left_on=["date"],right_on= ["date"])#display only matching record
```

```python
transactions.merge(oil,how ="left",left_on=["date"],right_on= ["date"])#retrieve all the record from 
                                                                #transactions records and only the matching record from oil
```
