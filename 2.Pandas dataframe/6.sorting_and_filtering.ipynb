### query() method lets you use SQL like syntax to filter dataframes.


```python
retail.loc[(retail["family"].isin(["CLEANING"])) &(retail["sales"]>100)]
```

```python
retail
```

```python
#the above code can be written using query() method

retail.query("family in ['CLEANING'] and sales >100")
```

```python
avg_sales=retail["sales"].mean()
```

```python
avg_sales
```

```python
retail.query("family in ['CLEANING','DAIRY'] and sales >@avg_sales")
```

```python
#sorting dataframe by indices

condition = retail["family"].isin(["CLEANING",'DAIRY',"DELI"])
```

```python
sample =retail[condition].sample(10,random_state=111)
```

```python
sample.sort_index(ascending=False)
```

```python
sample.sort_values("sales")
```

```python
#in this sample sales column in sorted ascending
#family column in not ascending
sample.sort_values(["sales","family"],ascending=[True,False])
```
