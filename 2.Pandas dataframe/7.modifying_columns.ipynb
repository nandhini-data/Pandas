```python
#arithmetic and boolean column creation
#by using series operation

retail_df =pd.read_csv("retail_2016_2017.csv")
```

```python
retail_df
```

```python
sample_df = retail_df.sample(20)
```

```python
sample_df
```

```python
sample_df["tax_amount"]=sample_df["sales"] * 0.05
```

```python
sample_df
```

```python
sample_df["total"] = sample_df["sales"] + sample_df["tax_amount"]
```

```python
sample_df
```

```python
sample_df["family"].value_counts()
```

```python
#create boolean column by assigning the logical test

sample_df["tax_category"] = sample_df["family"] !="BABY CARE"
```

```python
sample_df
```

```python
sample_df
```

```python
#numpy select() function let you create column based on multiple condition

conditions =[
    (sample_df["family"]=="HOME CARE") & (sample_df["date"]=="2016-07-23"),
    (sample_df["family"]=="PET SUPPLIES") & (sample_df["date"]=="2016-05-04"),
    (sample_df["store_nbr"] >28)
]

choice =["winter clearance","christmas","new store"]

sample_df["sale name"] =np.select(conditions,choice,default="no sale")

sample_df
```

```python
small_sample = sample_df.head(5)
small_sample
```

```python
#map() method maps values to a column or an entire dataframe
#you can pass a dictionary with existing value as the key and 
#new value as value.

mapping_dict = {"PET SUPPLIES":"animals","HOME CARE":"living","LAWN AND GARDEN":"gardening",
              "GROCERY I":"living","MEATS":"living"}

small_sample["family_category"]=small_sample["family"].map(mapping_dict)
small_sample
```

```python
assign_df = small_sample.iloc[:,0:5]
```
