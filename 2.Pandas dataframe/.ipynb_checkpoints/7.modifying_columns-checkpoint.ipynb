# <center>Modifying columns</center>


```python
import pandas as pd
import numpy as np
```

```python
product_df =pd.DataFrame({'product':["dairy","dairy","vegetables","fruits"],
                         'cost':[3,5,69,10]})
```

```python
#renaming columns

product_df
```

```python
product_df.columns =["Product_name","Price"]
```

```python
product_df
```

```python
product_df.columns = [ x.upper() for x in product_df.columns]
```

```python
product_df
```

```python
#using .rename() method to rename the columns

product_df.rename(columns ={'PRODUCT_NAME':'Product_list','PRICE':'cost_of_product'})
```

```python
#reordering columns

product_df.reindex(labels=['PRICE','PRODUCT_NAME'],axis=1)
```

```python
#arithmetic and boolean column creation
#by using series operation

retail_df =pd.read_csv("retail_2016_2017.csv")
```

```python
retail_df
```

```python
sample_df = retail_df.sample(20)
```

```python
sample_df
```

```python
sample_df["tax_amount"]=sample_df["sales"] * 0.05
```

```python
sample_df
```

```python
sample_df["total"] = sample_df["sales"] + sample_df["tax_amount"]
```

```python
sample_df
```

```python
sample_df["family"].value_counts()
```

```python
#create boolean column by assigning the logical test

sample_df["tax_category"] = sample_df["family"] !="BABY CARE"
```
