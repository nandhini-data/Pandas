# <center>Aggregating Groups</center>

The .agg() method let you perform multiple aggregation on "groupby" object.

```python
import pandas as pd
```

```python
retail =pd.read_csv("retail_2016_2017.csv")
```

```python
retail.head()
```

### groupby()

```python
#.agg() method perform the aggregation on all compatible columns

mini_retail =(retail.groupby(["store_nbr","family"]).
            agg({'sales':['sum',"mean"],'onpromotion':['min','max']}))

#here we perform sum and mean for sales column
mini_retail

```

### Inorder to avoid multi-index columns you can name the columns upon creation.

```python
retail_small =(retail.groupby(["family","store_nbr"])
                .agg(sales_sum = ("sales","sum"),sales_avg = ("sales","mean")))
```

```python
#single column index
retail_small
```

```python
league = pd.read_excel("premier_league_games_full.xlsx")
```

```python
league.head()
```

```python
(league.groupby(["season","HomeTeam"],as_index=False)
 .agg({'HomeGoals':['sum','mean'],
       'AwayGoals':['sum','mean']}))
```

```python
(league.groupby(["season","HomeTeam"],as_index=False)
 .agg(homegoal_sum = ('HomeGoals','sum'),
      awaygoal_sum = ('AwayGoals','sum'),homegoal_mean = ('HomeGoals','mean'),
      awaygoal_mean= ('AwayGoals','mean')))
```

### transform()

```python
#.transform() method can be used to perform aggregation without reshaping

retail
```

```python
retail.assign(new_store_sale = (retail.groupby("store_nbr")["sales"].transform('sum')))
```

```python
league
```

```python
league = league.assign(avg_team_goals = league.groupby("HomeTeam")["HomeGoals"].transform("mean"))
```
