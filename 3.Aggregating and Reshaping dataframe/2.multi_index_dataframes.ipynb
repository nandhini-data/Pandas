```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv("retail_2016_2017.csv")
```

```python
retail.head()
```

```python
retail_sum = retail.groupby(["family","store_nbr"])[["sales"]].sum().round()
```

```python
retail_sum
```

```python
#index of dataframe generally stored as a list of tuples,with an item for 
#each layer of the index
retail_sum.index
```

```python
#all rows of automotive is displayed
retail_sum.loc["AUTOMOTIVE"].head()
```

```python
#all rows from automotive to beauty is displayed
retail_sum.loc["AUTOMOTIVE":"BEAUTY"]
```

```python
#we can access the row through outer and inner indices
retail_sum.loc[("AUTOMOTIVE",12)]
```

```python
team =pd.read_excel("premier_league_games_full.xlsx")
```

```python
team
```

```python
agg_team =team.groupby(["season","HomeTeam"])[["HomeGoals"]].sum()
```

```python
agg_team.head()
```

```python
agg_team.loc[("2008/2009","Bolton Wanderers")]
```

```python
#multiple aggregation

agg_sum_mean = team.groupby(["season","HomeTeam"]).agg({"HomeGoals":["sum","mean"]})
```

```python
agg_sum_mean
```

```python
reset_agg_team = agg_sum_mean.reset_index()
```
