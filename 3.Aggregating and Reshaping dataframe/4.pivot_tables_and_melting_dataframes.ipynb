# <center>Pivot table and Melting dataframes</center>

```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
retail =pd.read_csv("retail_2016_2017.csv")
retail.head()
```

```python
small_retail = retail.query('store_nbr in(2,3,4)')
```

```python
small_retail
```

```python
small_retail.pivot_table(index = "family",columns = "store_nbr",values = "sales",aggfunc="mean").round(2)
```

```python
small_retail
```

```python
small_retail.pivot_table(index=('family','store_nbr'),
                        aggfunc=('sum','mean'),values=('sales','onpromotion'))
```

### Heatmaps

You can style a dataframe based on its value to create a heatmap

```python
import matplotlib.pyplot as plt
```

```python
sns.set(style='ticks', palette='RdBu')
```

```python
heatmap_df = small_retail.pivot_table(index = "family",
                          columns = "store_nbr",values = "sales",
                          aggfunc="mean")
cm = sns.light_palette("green",as_cmap =True)
heatmap_df .style.background_gradient(cmap=cm)
```

###  Melting dataframe

```python
premier =pd.read_excel("premier_league_games_full.xlsx")
```

```python
premier.head()
```

```python
pivot_premier = (premier.query("HomeTeam in ['Arsenal','Chelsea','Manchester United','Sunderland']")
 .pivot_table(index="HomeTeam",columns="season",values="HomeGoals",aggfunc="mean"))
```

```python
pivot_premier
```

```python
#.melt() is used for melting the above dataframe

pivot_premier.reset_index().melt(id_vars ="HomeTeam",
                    value_vars =("2008/2009","2009/2010","2010/2011"),
                                 value_name="avg_goals")
```
